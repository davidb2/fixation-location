```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
G = nx.cycle_graph(np.arange(10)+1, nx.DiGraph)
for u, v in G.edges():
  G.add_edge(v, u)

pos = nx.layout.kamada_kawai_layout(G)
```

```python
connectionstyle = "arc3,rad=0.2"
colors = sns.color_palette('Greens_d', n_colors=10)
nx.draw(
  G,
  pos=pos,
  connectionstyle=connectionstyle,
  arrows=True,
  node_color=[colors[i] for i in range(10)],
  width=.125*15,
  node_size=1000,
  arrowsize=10*4,
  with_labels=True,
)
plt.savefig(f'../../figs/figure_4_bidirected_cycles/presi-cycle-bidirected.eps', transparent=True, dpi=300, bbox_inches='tight')
```
