```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
G = nx.random_regular_graph(n=50, d=10)

pos = nx.layout.spring_layout(G)
connectionstyle = "arc3,rad=0.2"
colors = sns.color_palette('tab20', n_colors=20)
nx.draw(
  G,
  pos=pos,
  connectionstyle=connectionstyle,
  arrows=True,
  node_color=colors[8],
  width=.125*15,
  node_size=100*2,
)
plt.savefig(f'../../figs/figure_6_random_regular/rr.eps', transparent=True, dpi=300, bbox_inches='tight')
```
